## Load Books

```python
import os
#Get book data
books_text = [b for b in os.scandir("data") if '.txt' in b.name]
books_text
```

```python
NLP = spacy.load("en_core_web_sm")
text_in_book = open(books_text[1]).read()
book_data = NLP(text_in_book)
```

```python
displacy.render(book_data[0:1000],style="ent")
```

```python
# Removing later part of the name as the series uses name to describe where they are form
# While reading the book it would not always mention full names
import re
name_List['character'] = name_List['character'].apply(lambda x: re.sub("[\(].*?[\)]", "", x))
name_List['firstname'] = name_List['character'].apply(lambda x: x.split(' ',1)[0])
name_List[100:160]
```

```python
linear_node_df = []
for line in book_data.sents:
    node_list = [node.text for node in line.ents]
    linear_node_df.append({"sentance": line, "node": node_list})

linear_node_df = pd.DataFrame(linear_node_df)
linear_node_df
```

```python
linear_node_df[0:600]
```

```python
def remove_non_character(nodes, characters):
    return [node for node in nodes
            if node in list(characters.firstname)
            or node in list(characters.character)]
```

```python
linear_node_df["node_character"]= linear_node_df["node"].apply(lambda x: remove_non_character(x, name_List))
linear_node_df[60:120]              
```

```python
linear_node_df_updated = linear_node_df[linear_node_df["node_character"].map(len)>0]
linear_node_df_updated[0:50]
```

```python
linear_node_df_updated['node_character'] = linear_node_df_updated['node_character'].apply(lambda x: [item.split()[0]for item in x])
#pd.reset_option('^display.', silent = True)
linear_node_df_updated.head(50)
```

```python
## Relationships between characters used to making nodes
```

```python
window_size = 5
relationship = []
for i in range(linear_node_df_updated.index[-1]):
    end = min(i+5, linear_node_df_updated.index[-1])
    char_list = sum((linear_node_df_updated.loc[i:end].node_character), [])
    # Removing Duplicates
    char_list_unique = [char_list[i] for i in range(len(char_list)) if (i==0) or char_list[i] != char_list[i-1]]
    if len(char_list_unique) > 1:
        for idx, a in enumerate(char_list_unique[:-1]):
            b = char_list_unique[idx +1]
            relationship.append({"source": a, "target":b})
```

```python
relationship_df = pd.DataFrame(relationship)
relationship_df
```

```python
#Sort source with target and target with source for same characters
#As We are making non directional edges
relationship_df = pd.DataFrame(np.sort(relationship_df.values, axis =1), columns = relationship_df.columns)
relationship_df
```

```python
#Combining relationships
relationship_df["count"] = 1
relationship_df = relationship_df.groupby(["source","target"], sort = False, as_index=False).sum()
relationship_df.head(50)
```

```python
#Graph analysis and visulization
G = nx.from_pandas_edgelist(relationship_df, source = "source", target= "target", edge_attr = "count", create_using = nx.Graph())
```

```python
##Graph Visulization
from pyvis.network import Network
network_graph = Network(notebook = True, width="1000px", height="800px", bgcolor= "#222222", font_color="grey")
node_degree = dict(G.degree)
nx.set_node_attributes(G, node_degree, "size")
network_graph.from_nx(G)
network_graph.show("Witcher.html")
```

```python
dict(node_degree)
# Based o Degree centrality
degree_dict = nx.degree_centrality(G)
degree_dict
degree_df = pd.DataFrame.from_dict(degree_dict, orient = "index", columns=['centrality'])
degree_df.sort_values('centrality', ascending = False)[0:19].plot(kind="bar")
```

```python
# Betweenness centrality
betweenness_dict = nx.betweenness_centrality(G)
betweenness_dict
betweenness_df = pd.DataFrame.from_dict(betweenness_dict, orient = "index", columns=['B_centrality'])
betweenness_df.sort_values('B_centrality', ascending = False)[0:19].plot(kind="bar")
```

```python
#Closeness Centrality
closeness_dict = nx.closeness_centrality(G)
closeness_dict
closeness_df = pd.DataFrame.from_dict(closeness_dict, orient = "index", columns=['C_centrality'])
closeness_df.sort_values('C_centrality', ascending = False)[0:19].plot(kind="bar")
```

```python
#Saving centrality
nx.set_node_attributes(G, degree_dict, 'degree_centrality')
nx.set_node_attributes(G, betweenness_dict, 'betweenness_centrality')
nx.set_node_attributes(G, closeness_dict, 'closeness_centrality')
```

```python
import community as cl
```
